```python
# derste aldığım notlar 27

import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas") # x eksen adı
plt.ylabel("Kilo") #y eksen adı
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği") # tüm grafiğin başlığı
```

 yukarıdaki plt.plot(numpyYasListesi,numpyKiloListesi,"g") komutunda ilk verdiğimiz numpyYasListesi x eksenine yazıldı numpyKiloListesi ise y eksenine yazıldı , ayrıca g stringini green renkte çizmesi içi yazdık yani r=red, y=yellow

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 **3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b--") # kesikli grafik
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-") # yıldız grafik
```

```python
plt.subplot(1,2,1) #1 sıra olacak, 2 kolon olacak , şuan 1.grafiği çiziyorum
plt.plot(numpyDizisi1,numpyDizisi2, "r*-")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şuan 2.grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1, "g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
 
    #aşağıda subplot kullanmayıpp içi içe grafik yapıcaz

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) 
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3]) # 0.5 ten sonraki birinci 0.3 uzunluk ikincisi genişlik belirler 

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure, benimEksenler) = plt.subplots()
```

```python
benimFigure
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
    
plt.tight_layout()
```

```python
yeniFigure = plt.figure()
# plt.figure(dpi=150) gibi değerler koyup ne olduğuna bak

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 **2, label="numpyDizisi **2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 **3, label="numpyDizisi **3")
yeniEksen.legend()
```

```python
yeniFigure.savefig("benimfigure.png" ,dpi=200) # png olarak bunu bilgisayara kaydettik, dpi kalitesini arttırıyo ve boyutunu
```
