```python
# derste aldığım notlar 28

import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 * numpyDizisi1
```

```python
numpyDizisi2
```

```python
(benimFigur, benimEksen) = plt.subplots()
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, color = "#CC5C1C") #üstteki komutlarla aynı bölüm içine yazılmalı, color rgb to hex
benimEksen.plot(numpyDizisi2,numpyDizisi1, "b")
```

```python
(yeniFigur , yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="blue", linewidth=5.0) # linewidth çizginin kalınlığını ayarlar
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color="yellow", linewidth=2.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color ="#CC5C1C", linestyle= "-." )
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color ="#CC5C1C", linestyle= ":" )
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color ="#CC5C1C", linestyle= "--" )

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color ="#000000", linestyle= "-." ,marker="o" ,markersize = 4,markerfacecolor="red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color ="#000000", linestyle= "-." ,marker="+" ,markersize = 8)
```
